```python
import cv2 as cv
import numpy as np
from utils import show_img
```

## 1.Read

```python
img = cv.imread('../VisualMaterials/Photos/lady.jpg')
img
```

```python
img.shape
```

## 2. Show to image

```python
cv.imshow('Woman', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

## 3. Converting to grayscale

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
show_img(gray, "gray")
```

## 4. Blur

```python
blur = cv.GaussianBlur(img, (5,5), cv.BORDER_DEFAULT)
show_img(blur, "blur")
```

## 5. Edge Cascade

```python
canny = cv.Canny(blur, 125, 175)
show_img(canny, "canny_edges")
```

## 6. Dilating

```python
dilated = cv.dilate(canny, (7,7), iterations=5)
show_img(dilated, "dilated_canny")
```

```python
dilated = cv.dilate(gray, (15,15), iterations=5)
show_img(dilated, "dilated_gray")
```

## 7. Eroding

```python
eroded = cv.erode(canny, (5,5), iterations=5)
show_img(eroded, "eroded")
```

```python
eroded_gray = cv.erode(gray, (3,3), iterations=3)
show_img(eroded_gray, "eroded_gray")
```
